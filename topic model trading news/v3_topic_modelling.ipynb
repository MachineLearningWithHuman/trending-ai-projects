Topic modeling is a type of statistical modeling used in natural language processing (NLP) and machine learning to discover abstract "topics" within a collection of documents. It is an unsupervised learning technique, meaning it does not require labeled data. Instead, it identifies patterns and structures in the text by analyzing word co-occurrence and distribution.

**Key Concepts:**

---



**Topics:** These are clusters of words that frequently occur together and represent a common theme or subject. For example, in a collection of news articles, topics might include "sports," "politics," or "technology."

**Document-Topic Distribution:** Each document is represented as a mixture of topics. For instance, a document might be 60% about "sports" and 40% about "politics."

**Topic-Word Distribution:** Each topic is represented as a distribution over words. For example, the "sports" topic might include words like "football," "score," and "team."

**Common Algorithms:**

---



**Latent Dirichlet Allocation (LDA):** The most popular topic modeling algorithm. It assumes that documents are generated by randomly mixing topics, and topics are generated by randomly mixing words.

**Non-Negative Matrix Factorization (NMF):** A dimensionality reduction technique that can also be used for topic modeling by factorizing the term-document matrix.

**Latent Semantic Analysis (LSA):** Uses singular value decomposition (SVD) to reduce the dimensionality of the term-document matrix and identify latent topics.

**Correlated Topic Model (CTM):** An extension of LDA that allows topics to be correlated.

**Applications:**

---



Document Clustering: Grouping similar documents together.

Information Retrieval: Improving search engines by understanding the context of documents.

Content Recommendation: Suggesting articles or products based on topic similarity.

Trend Analysis: Identifying emerging topics in large text corpora over time. \

Steps in Topic Modeling:

---



Preprocessing: Clean and prepare the text data (e.g., tokenization, removing stop words, stemming/lemmatization).

Vectorization: Convert text into a numerical format (e.g., using TF-IDF or word embeddings).

Model Training: Apply a topic modeling algorithm (e.g., LDA) to the vectorized data.

Interpretation: Analyze the resulting topics and their distributions across documents.

Topic modeling is a powerful tool for uncovering hidden structures in text data, making it invaluable for tasks like content analysis, summarization, and knowledge discovery.

# Latent Dirichlet Allocation (LDA)

---

Latent Dirichlet Allocation (LDA) is a generative probabilistic model used for topic modeling in natural language processing (NLP). It is one of the most popular algorithms for discovering latent (hidden) topics in a collection of documents. LDA assumes that documents are composed of a mixture of topics, and each topic is characterized by a distribution of words.


---


**Key Ideas Behind LDA:**


---


Generative Process:

LDA assumes that documents are generated in a probabilistic way:

For each document, a distribution over topics is randomly chosen.

For each word in the document, a topic is randomly selected from the document's topic distribution.

A word is then randomly chosen from the topic's word distribution.


---


Dirichlet Priors:

LDA uses Dirichlet distributions as priors for the topic distributions (per document) and word distributions (per topic). These priors ensure that the distributions are sparse, meaning a document is likely to focus on a few topics, and a topic is likely to focus on a few words.


---


Latent Variables:

The topics are "latent" because they are not directly observed. Instead, they are inferred from the observed words in the documents.


---

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from wordcloud import WordCloud
import gensim
from gensim import corpora
import pyLDAvis.gensim_models as gensimvis
import pyLDAvis

# Download NLTK resources
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
# Download the missing 'punkt_tab' resource
nltk.download('punkt_tab') # This line was added to fix the error

# Load financial news dataset (sample)
news_data = [
    "Tesla stock surges after strong Q4 earnings report",
    "Federal Reserve raises interest rates to combat inflation",
    "Apple unveils new AI-driven iPhone features",
    "Stock market dips as economic uncertainty looms",
    "Bitcoin hits new all-time high amid institutional buying",
    "Amazon expands cloud computing services to new markets",
    "Nvidia sees record sales as AI chip demand skyrockets",
    "Regulatory crackdown on cryptocurrency exchanges spooks investors"
]

df = pd.DataFrame(news_data, columns=["headline"])

# Preprocessing function
def preprocess_text(text):
    lemmatizer = WordNetLemmatizer()
    stop_words = set(stopwords.words('english'))
    tokens = word_tokenize(text.lower())  # Convert to lowercase and tokenize
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word.isalnum()]  # Remove non-alphanumeric
    tokens = [word for word in tokens if word not in stop_words]  # Remove stopwords
    return tokens




````

````python
df["tokens"] = df["headline"].apply(preprocess_text)

df.head()
````

````python
# Create dictionary and corpus
dictionary = corpora.Dictionary(df["tokens"])
corpus = [dictionary.doc2bow(text) for text in df["tokens"]]

dictionary,corpus
````

````python
# Apply LDA
lda_model = gensim.models.LdaModel(corpus, num_topics=3, id2word=dictionary, passes=15)

# Print Topics
topics = lda_model.print_topics(num_words=5)
for i, topic in topics:
    print(f"Topic {i}: {topic}")
````

````python
import seaborn as sns
import numpy as np
from gensim import corpora, models, similarities

# Create a topic similarity heatmap using get_topic_terms instead of similarity
def plot_topic_similarity(lda_model):
    num_topics = lda_model.num_topics
    topic_similarity = np.zeros((num_topics, num_topics))

    for i in range(num_topics):
        for j in range(num_topics):
            # Get the top words for each topic
            topic1_words = [word for word, prob in lda_model.get_topic_terms(i, topn=10)]  # Change topn as needed
            topic2_words = [word for word, prob in lda_model.get_topic_terms(j, topn=10)]  # Change topn as needed

            # Calculate similarity (e.g., using Jaccard similarity)
            common_words = set(topic1_words) & set(topic2_words)
            all_words = set(topic1_words) | set(topic2_words)
            topic_similarity[i, j] = len(common_words) / len(all_words) if all_words else 0  # Jaccard similarity

    sns.heatmap(topic_similarity, annot=True, cmap='YlGnBu')
    plt.xlabel('Topic ID')
    plt.ylabel('Topic ID')
    plt.title('Topic Similarity Heatmap')
    plt.show()

# Example: Plot topic similarity heatmap
plot_topic_similarity(lda_model)
````

````python
import networkx as nx
import matplotlib.pyplot as plt

# Create a topic-word network graph
def plot_topic_word_network(lda_model, num_words=5):
    G = nx.Graph()

    for topic_id in range(lda_model.num_topics):
        topic_words = lda_model.show_topic(topic_id, num_words)
        for word, weight in topic_words:
            G.add_edge(f"Topic {topic_id}", word, weight=weight)

    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', font_size=10, node_size=2000)
    plt.title('Topic-Word Network Graph')
    plt.show()

# Example: Plot topic-word network graph
plot_topic_word_network(lda_model)
````

````python
import pandas as pd
import matplotlib.pyplot as plt

# Plot topic evolution over time
def plot_topic_evolution(lda_model, corpus, timestamps):
    topic_weights_over_time = []
    for doc_id, timestamp in enumerate(timestamps):
        doc_topics = lda_model.get_document_topics(corpus[doc_id])
        topic_weights = {topic[0]: topic[1] for topic in doc_topics}
        topic_weights_over_time.append((timestamp, topic_weights))

    # Convert topic_weights_over_time to a list of dictionaries with numeric values
    topic_weights_list = []
    for timestamp, topic_weights in topic_weights_over_time:
        data_point = {'Timestamp': timestamp}  # Rename 0 to 'Timestamp'
        data_point.update(topic_weights)  # Add topic weights
        topic_weights_list.append(data_point)

    df = pd.DataFrame(topic_weights_list).set_index('Timestamp')  # Set 'Timestamp' as index
    df.plot(kind='area', stacked=True, figsize=(10, 6))
    plt.xlabel('Time')
    plt.ylabel('Topic Weight')
    plt.title('Topic Evolution Over Time')
    plt.show()

# Example: Plot topic evolution (assuming timestamps are available)
timestamps = pd.date_range(start='2023-01-01', periods=len(corpus), freq='D')
plot_topic_evolution(lda_model, corpus, timestamps)
````

````python
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

# Plot hierarchical clustering of topics
def plot_topic_clustering(lda_model):
    topic_embeddings = lda_model.get_topics()
    Z = linkage(topic_embeddings, method='ward')

    plt.figure(figsize=(10, 5))
    dendrogram(Z, labels=[f"Topic {i}" for i in range(lda_model.num_topics)])
    plt.xlabel('Topics')
    plt.ylabel('Distance')
    plt.title('Hierarchical Clustering of Topics')
    plt.show()

# Example: Plot topic clustering
plot_topic_clustering(lda_model)
````

````python
import networkx as nx
import plotly.graph_objects as go

# Create a topic-word network graph using Plotly
def plot_topic_word_network_plotly(lda_model, num_words=5):
    # Create a graph
    G = nx.Graph()

    # Add nodes and edges for topics and words
    for topic_id in range(lda_model.num_topics):
        topic_words = lda_model.show_topic(topic_id, num_words)
        for word, weight in topic_words:
            G.add_edge(f"Topic {topic_id}", word, weight=weight)

    # Get positions for the nodes using a layout algorithm
    pos = nx.spring_layout(G)

    # Extract node positions
    node_x = []
    node_y = []
    for node in G.nodes():
        x, y = pos[node]
        node_x.append(x)
        node_y.append(y)

    # Create edge traces
    edge_traces = []
    for edge in G.edges():
        x0, y0 = pos[edge[0]]
        x1, y1 = pos[edge[1]]
        edge_trace = go.Scatter(
            x=[x0, x1, None],  # Add None to create a gap between edges
            y=[y0, y1, None],
            line=dict(width=1, color='gray'),
            hoverinfo='none',
            mode='lines'
        )
        edge_traces.append(edge_trace)

    # Create node traces
    node_trace = go.Scatter(
        x=node_x,
        y=node_y,
        mode='markers+text',
        text=list(G.nodes()),
        textposition="top center",
        marker=dict(
            size=20,
            color='lightblue',
            line=dict(width=2, color='black')
        ),
        hoverinfo='text'
    )

    # Create the figure
    fig = go.Figure(data=edge_traces + [node_trace])

    # Update layout
    fig.update_layout(
        title='Topic-Word Network Graph',
        showlegend=False,
        hovermode='closest',
        xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
        yaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
        margin=dict(l=20, r=20, t=40, b=20)
    )

    # Show the figure
    fig.show()

# Example: Plot topic-word network graph using Plotly
plot_topic_word_network_plotly(lda_model)
````

**Non-negative Matrix Factorization (NMF)** is a **dimensionality reduction** and **factorization technique** commonly used in machine learning and data analysis. It decomposes a non-negative matrix into two lower-dimensional non-negative matrices, which can be interpreted as parts-based representations of the original data. NMF is particularly popular in **text mining**, **image processing**, and **recommender systems**.

---

2. **Optimization**:
   The optimization problem is solved using iterative algorithms such as:
   - **Multiplicative Update Rules**: Updates \( W \) and \( H \) iteratively while preserving non-negativity.
   - **Alternating Least Squares (ALS)**: Alternately fixes \( W \) and \( H \) and solves for the other.

---

### **Interpretation**
- **Basis Matrix (\( W \))**:
  - Represents the "parts" or "components" of the data.
  - In text mining, each column of \( W \) corresponds to a **topic**, and the values represent the contribution of each word to the topic.
  - In image processing, each column of \( W \) corresponds to a **basis image**.

- **Coefficient Matrix (\( H \))**:
  - Represents the weights or activations of the components for each data point.
  - In text mining, each row of \( H \) corresponds to a **document**, and the values represent the contribution of each topic to the document.
  - In image processing, each row of \( H \) corresponds to a **pixel**, and the values represent the contribution of each basis image to the pixel.

---

### **Applications of NMF**
1. **Text Mining and Topic Modeling**:
   - NMF is used to discover latent topics in a corpus of documents.
   - Each column of \( W \) represents a topic, and each row of \( H \) represents the topic distribution for a document.

2. **Image Processing**:
   - NMF is used for **image compression** and **feature extraction**.
   - Each column of \( W \) represents a basis image, and each row of \( H \) represents the weights for reconstructing the original image.

3. **Recommender Systems**:
   - NMF is used to factorize user-item interaction matrices to discover latent factors (e.g., user preferences and item features).

4. **Bioinformatics**:
   - NMF is used to analyze gene expression data and identify patterns or clusters.

---

### **Advantages of NMF**
1. **Parts-Based Representation**:
   - NMF provides a parts-based representation of the data, which is often more interpretable than other factorization methods (e.g., PCA).

2. **Non-Negativity**:
   - The non-negativity constraint ensures that the components and weights are additive, making the results easier to interpret.

3. **Scalability**:
   - NMF can handle large datasets efficiently, especially with sparse matrices.

---

### **Limitations of NMF**
1. **Non-Convex Optimization**:
   - The objective function is non-convex, so the solution may depend on the initialization and converge to a local minimum.

2. **Choice of \( k \)**:
   - The number of components \( k \) must be chosen in advance, which can be challenging.

3. **Sensitive to Noise**:
   - NMF may not perform well on noisy data, as it assumes a clean, additive structure.

---

### **Example: Topic Modeling with NMF**
Here’s how NMF can be applied to topic modeling:

#### Step 1: Preprocess the Text Data
- Tokenize the text.
- Remove stopwords and punctuation.
- Create a document-term matrix \( V \), where each row represents a document and each column represents a word.

#### Step 2: Apply NMF
```python
from sklearn.decomposition import NMF
from sklearn.feature_extraction.text import TfidfVectorizer

# Example documents
documents = [
    "I love machine learning and data science",
    "Artificial intelligence is the future",
    "I enjoy hiking and outdoor activities",
    "Sports and fitness are important for health"
]

# Create TF-IDF document-term matrix
vectorizer = TfidfVectorizer(stop_words='english')
V = vectorizer.fit_transform(documents)

# Apply NMF
num_topics = 2
nmf = NMF(n_components=num_topics, random_state=42)
W = nmf.fit_transform(V)  # Document-topic matrix
H = nmf.components_      # Topic-term matrix

# Display topics
feature_names = vectorizer.get_feature_names_out()
for topic_idx, topic in enumerate(H):
    print(f"Topic {topic_idx}:")
    print(" ".join([feature_names[i] for i in topic.argsort()[:-10:-1]]))
```

#### Step 3: Interpret the Results
- \( W \) (Document-Topic Matrix): Shows the contribution of each topic to each document.
- \( H \) (Topic-Term Matrix): Shows the contribution of each word to each topic.

---

### **Comparison with Other Methods**
1. **NMF vs. PCA**:
   - PCA allows negative values, which can make interpretation difficult.
   - NMF enforces non-negativity, leading to more interpretable results.

2. **NMF vs. LDA**:
   - LDA is a probabilistic model, while NMF is a deterministic matrix factorization method.
   - NMF is often faster and easier to implement but may not capture complex relationships as well as LDA.

---

### **Conclusion**
NMF is a powerful and interpretable technique for discovering latent structures in non-negative data. Its parts-based representation makes it particularly useful for applications like topic modeling, image processing, and recommender systems. However, care must be taken in choosing the number of components and handling noisy data.

````python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import NMF
from wordcloud import WordCloud

# Load sample earnings call data
earnings_calls = [
    "Our revenue growth was strong this quarter, driven by AI expansion and cloud computing demand.",
    "We faced regulatory challenges, but our long-term outlook remains positive with new product launches.",
    "Increased competition in the market has put pressure on profit margins and growth forecasts.",
    "Strong consumer demand has boosted our earnings, with record-breaking sales in the holiday season.",
    "We continue to invest heavily in AI and automation to enhance operational efficiency and profitability.",
    "The central bank's interest rate hike has led to increased borrowing costs, affecting financial performance."
]

df = pd.DataFrame(earnings_calls, columns=["transcript"])

# Preprocessing function
nltk.download("stopwords")
nltk.download("punkt")
nltk.download("punkt_tab")

def preprocess_text(text):
    stop_words = set(stopwords.words("english"))
    tokens = word_tokenize(text.lower())
    tokens = [word for word in tokens if word.isalnum() and word not in stop_words]
    return " ".join(tokens)

df["processed"] = df["transcript"].apply(preprocess_text)

# Convert text data to TF-IDF features
vectorizer = TfidfVectorizer(max_features=1000)
X = vectorizer.fit_transform(df["processed"])

# Apply NMF
num_topics = 3
nmf_model = NMF(n_components=num_topics, random_state=42)
nmf_topics = nmf_model.fit_transform(X)

# Get top words per topic
feature_names = vectorizer.get_feature_names_out()
top_words_per_topic = {}

for topic_idx, topic in enumerate(nmf_model.components_):
    top_words = [feature_names[i] for i in topic.argsort()[-10:]]
    top_words_per_topic[f"Topic {topic_idx}"] = top_words

# Print Topics
for topic, words in top_words_per_topic.items():
    print(f"{topic}: {', '.join(words)}")

# Heatmap Visualization
plt.figure(figsize=(10, 6))
sns.heatmap(nmf_topics, annot=True, cmap="Blues", xticklabels=[f"Topic {i}" for i in range(num_topics)])
plt.xlabel("Topics")
plt.ylabel("Earnings Call Index")
plt.title("Topic Distribution Across Earnings Calls")
plt.show()
````

````python
# Visualization 5: Network Graph of Topics and Words (Using Plotly)
import networkx as nx
def plot_topic_word_network_plotly(nmf_model, top_words_per_topic):
    G = nx.Graph()
    for topic, words in top_words_per_topic.items():
        for word in words:
            G.add_edge(topic, word)

    pos = nx.spring_layout(G)
    edge_traces = []
    for edge in G.edges():
        x0, y0 = pos[edge[0]]
        x1, y1 = pos[edge[1]]
        edge_trace = go.Scatter(
            x=[x0, x1, None], y=[y0, y1, None],
            line=dict(width=1, color='gray'),
            hoverinfo='none',
            mode='lines'
        )
        edge_traces.append(edge_trace)

    node_x = [pos[node][0] for node in G.nodes()]
    node_y = [pos[node][1] for node in G.nodes()]
    node_trace = go.Scatter(
        x=node_x, y=node_y,
        mode='markers+text',
        text=list(G.nodes()),
        textposition="top center",
        marker=dict(size=20, color='lightblue', line=dict(width=2, color='black')),
        hoverinfo='text'
    )

    fig = go.Figure(data=edge_traces + [node_trace])
    fig.update_layout(
        title='Topic-Word Network Graph',
        showlegend=False,
        hovermode='closest',
        xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
        yaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
        margin=dict(l=20, r=20, t=40, b=20)
    )
    fig.show()

plot_topic_word_network_plotly(nmf_model, top_words_per_topic)
````

````python
from bertopic import BERTopic
from sentence_transformers import SentenceTransformer
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import umap # Import the umap module
from hdbscan import HDBSCAN

# Load sample social media discussions
reddit_posts = [
    "Tesla stock is going to the moon after the latest AI breakthrough!",
    "The Fed's rate hikes are killing the stock market, bearish sentiment rising.",
    "Nvidia's dominance in the AI chip market is insane, huge growth potential.",
    "Crypto markets are volatile, but institutions keep buying Bitcoin.",
    "Short squeeze incoming on GameStop, retail traders piling in again!",
    "Big money is rotating into defensive stocks, recession fears looming."
]

df = pd.DataFrame(reddit_posts, columns=["post"])

# Load sentence transformer model
embedding_model = SentenceTransformer("all-MiniLM-L6-v2")

# Create embeddings
embeddings = embedding_model.encode(df["post"].tolist(), show_progress_bar=True)

# Apply BERTopic with adjusted parameters
# Lower min_samples and adjust n_neighbors and min_dist for UMAP
topic_model = BERTopic(
    umap_model=umap.UMAP(n_neighbors=3, n_components=2, min_dist=0.1),  # Increased n_neighbors and min_dist
    hdbscan_model=HDBSCAN(min_samples=2, min_cluster_size=2)  # Reduced min_samples and min_cluster_size
)
topics, _ = topic_model.fit_transform(df["post"].tolist(), embeddings)

# Get the number of unique topics (excluding -1 for outliers)
num_topics = len(set(topics)) - (1 if -1 in topics else 0)
print(num_topics)
# Visualize topic distribution
topic_model.visualize_barchart(top_n_topics=min(num_topics,5))
````

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.sentiment import SentimentIntensityAnalyzer
from wordcloud import WordCloud
import gensim
from gensim import corpora
import pyLDAvis.gensim_models as gensimvis
import pyLDAvis

# Download NLTK resources
nltk.download("stopwords")
nltk.download("punkt")
nltk.download("vader_lexicon")

# Load sample financial news data
news_articles = [
    "Tesla stock soars after massive earnings beat, investors bullish on AI prospects.",
    "Economic slowdown fears rise as Fed signals further interest rate hikes.",
    "Apple announces major product innovation, boosting investor confidence.",
    "Stock market dips amid geopolitical tensions and rising oil prices.",
    "Bitcoin surges to new highs as institutional adoption accelerates.",
    "Bear market fears deepen as consumer spending weakens, GDP growth slows."
]

df = pd.DataFrame(news_articles, columns=["article"])

# Preprocessing function
def preprocess_text(text):
    stop_words = set(stopwords.words("english"))
    tokens = word_tokenize(text.lower())
    tokens = [word for word in tokens if word.isalnum() and word not in stop_words]
    return tokens

df["tokens"] = df["article"].apply(preprocess_text)

# Create dictionary and corpus
dictionary = corpora.Dictionary(df["tokens"])
corpus = [dictionary.doc2bow(text) for text in df["tokens"]]

# Apply LDA
lda_model = gensim.models.LdaModel(corpus, num_topics=3, id2word=dictionary, passes=15)

# Print Topics
topics = lda_model.print_topics(num_words=5)
topic_words = {}

for i, topic in topics:
    print(f"Topic {i}: {topic}")
    words = [word.split("*")[-1].replace('"', "").strip() for word in topic.split("+")]
    topic_words[f"Topic {i}"] = words

# Sentiment Analysis
sia = SentimentIntensityAnalyzer()

def get_topic_sentiment(words):
    scores = [sia.polarity_scores(word)["compound"] for word in words]
    avg_score = np.mean(scores)
    return "Bullish" if avg_score > 0 else "Bearish"

sentiments = {topic: get_topic_sentiment(words) for topic, words in topic_words.items()}

# Convert to DataFrame for visualization
sentiment_df = pd.DataFrame(sentiments.items(), columns=["Topic", "Sentiment"])

# Visualization 1: Bar Plot of Sentiments
plt.figure(figsize=(8, 5))
sns.countplot(x="Sentiment", data=sentiment_df, palette={"Bullish": "green", "Bearish": "red"})
plt.title("Bullish vs. Bearish Topics")
plt.show()

# Visualization 2: Pie Chart of Sentiment Distribution
sentiment_counts = sentiment_df["Sentiment"].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(sentiment_counts, labels=sentiment_counts.index, autopct="%1.1f%%", colors=["green", "red"])
plt.title("Sentiment Distribution of Topics")
plt.show()

# Visualization 3: Word Clouds for Bullish vs Bearish
bullish_words = [word for topic, words in topic_words.items() if sentiments[topic] == "Bullish" for word in words]
bearish_words = [word for topic, words in topic_words.items() if sentiments[topic] == "Bearish" for word in words]

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

wordcloud_bullish = WordCloud(width=600, height=400, background_color="white", colormap="Greens").generate(" ".join(bullish_words))
axes[0].imshow(wordcloud_bullish, interpolation="bilinear")
axes[0].set_title("Bullish Topics")
axes[0].axis("off")

wordcloud_bearish = WordCloud(width=600, height=400, background_color="white", colormap="Reds").generate(" ".join(bearish_words))
axes[1].imshow(wordcloud_bearish, interpolation="bilinear")
axes[1].set_title("Bearish Topics")
axes[1].axis("off")

plt.tight_layout()
plt.show()
````
